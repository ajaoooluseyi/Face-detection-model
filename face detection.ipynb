```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
cap_mavl = cv2.imread(r'C:\Users\AJAO SEYI\Desktop\ML\cpt.png')

roi = cap_mavl[30:170, 150:320]
roi = cv2.cvtColor(roi, cv2.COLOR_BGR2GRAY)
plt.imshow(roi, cmap = 'gray')
plt.axis('off')
```

```python
# Load Cascade filter 
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")
```

```python
# Create the face detecting function 
def detect_face(img):
    
    face_rects = face_cascade.detectMultiScale(img, 
                                               scaleFactor = 1.1,
                                               minNeighbors = 3)
    
    for (x, y, w, h) in face_rects:
        cv2.rectangle(img, (x, y), (x+w, y+h), (255, 255, 255), 3)
        
    return img

# Detect the face
roi_detected = detect_face(roi)
plt.imshow(roi_detected, cmap = 'gray')
plt.axis('off')
```

```python
pot = cv2.imread(r'C:\Users\AJAO SEYI\Desktop\ML\pot.jpg')
pot = cv2.cvtColor(pot, cv2.COLOR_BGR2RGB)

faces = detect_face(pot)
plt.imshow(faces, cmap = 'gray')
plt.axis('off')
```

```python
capture = cv2.VideoCapture(0)
num_faces = 0
new_dimension = (200, 200)

while True:
    response, frame = capture.read()
    if detect_face(frame) is not None:
        num_faces += 1

        (x,y,w,h), found_face = detect_face(frame)
        # resizing found face to new dimensions
        resized_face = cv2.resize(found_face, new_dimension)
        # Converting to grayscale again
        resized_face = cv2.cvtColor(resized_face, cv2.COLOR_BGR2GRAY)

        # Saving the training data
        path = 'trainingData/' + str(num_faces) + '.jpg'
        cv2.imwrite(path, resized_face)

        # To track number of images completed
        cv2.putText(resized_face, str(num_faces), (50,50), cv2.FONT_ITALIC, 1, (0,0,0), 2)
        cv2.rectangle(frame, (x, y), (x+w, y+h), (0,250,0), 2)
        cv2.imshow('Saving training data', frame)
        cv2.imshow("Found face", resized_face)

    else:
        print ("No face found")
        pass
    if cv2.waitKey(1) == 13 or  num_faces == 50:
        break

capture.release()
cv2.destroyAllWindows()

print ('Training data done')
```
